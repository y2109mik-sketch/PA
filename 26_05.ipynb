```python
#Projeto caixa eletrônico

#Você foi contratado para desenvolver um sistema simples de caixa eletrônico

#O sistema deverá funcionar através de um menu interativo utilizando "while"

#O programa deve iniciar com um um saldo de R$ 1000,00 e permitir que o usuário
#escolha diferentes operaçoes até decidir


saldo = 1000.00
opcao = 0

while opcao != 4:
  print("-------MENU-------")
  print("")
  print("Opção 1 - Consultar saldo")
  print("Opção 2 - Saque")
  print("Opção 3 - Depósito")
  print("Opção 4 - Sair")
  print("")


  opcao = int(input("Escolha a opção desejada: "))

  if opcao == 1:
    print(f"Seu saldo é de R$ {saldo}")

  elif opcao == 2:
    saque = float(input("Digite o valor do saque: "))

    if saque > saldo:
        print("Saldo insuficiente")

    else:
        saldo -= saque
        print(f"Saque de R$ {saque} realizado com sucesso")
        print(f"Seu saldo atual é de R$ {saldo}")

  elif opcao == 3:
        deposito = float(input("Digite o valor do depósito: "))
        saldo += deposito
        print(f"Depósito de R$ {deposito} realizado com sucesso")
        print(f"Seu saldo atual é de R$ {saldo}")

  elif opcao == 4:
      print("")
      print("Obrigado por usar o caixa eletrônico")
      break

  else:
        print("Opção inválida")
```
